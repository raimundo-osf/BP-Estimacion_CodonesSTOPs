```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo Excel
file_path = 'https://github.com/raimundo-osf/BounsPoint1/raw/main/resultados_stop_codons.xlsx'  # Ruta de archivo
df = pd.read_excel(file_path)

# Filtrado de columnas
df_filtered = df[['%G+C', 'Estimación 5']]

# Graficar el %G+C vs. la estimación 5
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_filtered['%G+C'], y=df_filtered['Estimación 5'])
plt.xlabel('%G+C')
plt.ylabel('Estimación 5')
plt.title('%G+C vs. Estimación 5')
plt.grid(True)
plt.show()

# Calcular la correlación entre %G+C y la estimación 5
correlacion_5 = df_filtered['%G+C'].corr(df_filtered['Estimación 5'])

print("Correlación entre %G+C y Estimación 5:", correlacion_5)
```
