```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo Excel
file_path = 'https://github.com/raimundo-osf/BounsPoint1/raw/main/resultados_stop_codons.xlsx'  # Ruta de archivo
df = pd.read_excel(file_path)

# Filtrado de columnas
df_filtered = df[['%G+C', 'Estimacin 5']]

# Graficar el %G+C vs. la estimacin 5
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_filtered['%G+C'], y=df_filtered['Estimacin 5'])
plt.xlabel('%G+C')
plt.ylabel('Estimacin 5')
plt.title('%G+C vs. Estimacin 5')
plt.grid(True)
plt.show()

# Calcular la correlacin entre %G+C y la estimacin 5
correlacion_5 = df_filtered['%G+C'].corr(df_filtered['Estimacin 5'])

print("Correlacin entre %G+C y Estimacin 5:", correlacion_5)
```
